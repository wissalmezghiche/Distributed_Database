**3. XML vers JSON**

*Q4 : écrivez le code Python qui convertit le fichier XML en code de 
langage JSON*

```python
import xml.etree.ElementTree as ET
import json

xml_file = "/kaggle/input/dataset/file.xml"


tree = ET.parse(xml_file)
root = tree.getroot()


def xml_to_dict(element):
    d = {}
    for child in element:
        
        if child:
            if child.tag in d:
                if isinstance(d[child.tag], list):
                    d[child.tag].append(xml_to_dict(child))
                else:
                    d[child.tag] = [d[child.tag], xml_to_dict(child)]
            else:
                d[child.tag] = xml_to_dict(child)
        else:
            
            d[child.tag] = child.text.strip() if child.text else None
        
        if child.attrib:
            d[child.tag] = {"text": d[child.tag], "attributes": child.attrib}
    return d


xml_dict = xml_to_dict(root)


json_data = json.dumps(xml_dict, indent=4)


print(json_data)
```

**4. XML vers JSON**

*Q5 : écrivez le code Python qui convertit le fichier JSON 
à côté d'un code de langage XML.*

```python
import json
import xml.etree.ElementTree as ET


with open('/kaggle/input/dataset/data.json') as f:
    json_data = json.load(f)


def dict_to_xml(parent, dictionary):
    for key, value in dictionary.items():
        if isinstance(value, dict):
            
            element = ET.SubElement(parent, key)
            dict_to_xml(element, value)
        elif isinstance(value, list):
            
            for item in value:
                dict_to_xml(parent, {key: item})
        else:
            
            element = ET.SubElement(parent, key)
            element.text = str(value)


root = ET.Element("root")


dict_to_xml(root, json_data)


tree = ET.ElementTree(root)

tree.write("data.xml")
```
