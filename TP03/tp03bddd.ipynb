```python
import folium
latitude , longitude  = 36.66406155322093, 4.912701598956947

m = folium.Map(location=[latitude, longitude], zoom_start=16)

folium.Circle(
    radius=200,
    location=[latitude, longitude],
    color="crimson",
    fill=False,
).add_to(m)

folium.Marker(
    location=[latitude, longitude],
    popup='Estin',
    tooltip='cliquez ici'
).add_to(m)

m
# m.save('map.html')
```
